```python
# 1. Transaction Volume and Amount over Time
daily_trend = df.groupby('date')['amount'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.lineplot(data=daily_trend, x='date', y='amount', color='blue')
plt.title('Total Daily Transaction Volume (AUD)', fontsize=15)
plt.xlabel('Date')
plt.ylabel('Total Amount')
plt.grid(True)
plt.savefig('daily_spending_trend.png')
```

```python
# 2. Spending by Gender
gender_spend = df[df['movement'] == 'debit'].groupby('gender')['amount'].mean().reset_index()
plt.figure(figsize=(8, 6))
sns.barplot(data=gender_spend, x='gender', y='amount', palette='viridis')
plt.title('Average Transaction Amount by Gender (Debits Only)', fontsize=15)
plt.ylabel('Average Transaction Amount (AUD)')
plt.savefig('gender_spending.png')
```

```python
# 3. Age Group Analysis
df['age_group'] = pd.cut(df['age'], bins=[0, 25, 35, 45, 55, 65, 100], labels=['<25', '25-34', '35-44', '45-54', '55-64', '65+'])
age_balance = df.groupby('age_group', observed=False)['balance'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(data=age_balance, x='age_group', y='balance', palette='magma')
plt.title('Average Account Balance by Age Group', fontsize=15)
plt.ylabel('Average Balance (AUD)')
plt.savefig('age_balance.png')
```

```python
# 4. Transaction Description Analysis
txn_type = df['txn_description'].value_counts()
plt.figure(figsize=(10, 8))
txn_type.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Transaction Types', fontsize=15)
plt.ylabel('')
plt.savefig('txn_distribution.png')
```

```python
# 5. Peak Spending Hours (Engagement)
hourly_activity = df.groupby('hour')['transaction_id'].count().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=hourly_activity, x='hour', y='transaction_id', color='salmon')
plt.title('Transaction Frequency by Hour of Day', fontsize=15)
plt.xlabel('Hour (24h Format)')
plt.ylabel('Number of Transactions')
plt.savefig('hourly_activity.png')

```

```python
# Calculate some summary stats for the text response
total_spend = df[df['movement'] == 'debit']['amount'].sum()
avg_salary = df[df['txn_description'] == 'PAY_REM']['amount'].mean()
top_state = df['merchant_state'].mode()[0]

print(f"Total Debit Spending: {total_spend}")
print(f"Average Salary Payment: {avg_salary}")
print(f"Most Active State: {top_state}")
```

```python
# Checking transaction descriptions to find salary tags
print(df['txn_description'].value_counts())
# Checking a few rows where movement is credit to see income types
print(df[df['movement'] == 'credit']['txn_description'].value_counts())
```

```python
#Day of week and Hour
df['day_name'] = df['date'].dt.day_name()
df['hour'] = pd.to_datetime(df['extraction']).dt.hour
# Plotting Day of Week Trend
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='day_name', order=day_order, palette='viridis')
plt.title('Transaction Frequency by Day of Week')
plt.savefig('day_freq.png')
```

```python
# Corrected code using seaborn's countplot
plt.figure(figsize=(8, 3)) # 'height=300' roughly maps to a height of 3 in matplotlib
sns.countplot(data=df, y='movement', palette='viridis')

plt.title('Movement')
plt.xlabel('Number of Transactions')
plt.ylabel('Movement Type')
plt.tight_layout()
plt.savefig('movement_distribution.png')
```

```python
# 5. Gender based average balance
gender_balance = df.groupby('gender')['balance'].mean().reset_index()
plt.figure(figsize=(6, 6))
sns.barplot(x='gender', y='balance', data=gender_balance)
plt.title('Average Account Balance by Gender')
plt.savefig('gender_balance.png')

```

```python
# Debit transactions only (Expenditure)
debit_df = df[df['movement'] == 'debit'].copy()

#Average Debit by Specific Age
age_avg_debit = debit_df.groupby('age')['amount'].mean().reset_index()

#Age GroupP
bins = [18, 25, 35, 45, 55, 65, 100]
labels = ['18-25', '26-35', '36-45', '46-55', '56-65', '65+']
debit_df['age_group'] = pd.cut(debit_df['age'], bins=bins, labels=labels, right=False)

age_group_summary = debit_df.groupby('age_group')['amount'].mean().reset_index()

#Visualization
plt.figure(figsize=(10, 6))
sns.barplot(data=age_group_summary, x='age_group', y='amount', palette='coolwarm')
plt.title('Average Expenditure (Debit Amount) by Age Group')
plt.ylabel('Average Amount (AUD)')
plt.show()
plt.savefig('average_amount.png') 
plt.show()                 
print(age_group_summary)
```
